# Adjusting Architecture by DCA on the Nodes of Neural Network

前言
---
> 最近，老板有一件事情要我去做。这件事情简单来说是这样子，有一个简单的全连接神经网络，首先进行预训练，然后用DCA计算不同层节点之间的关联，问题是，我们能否在原有的全连接网络基础上，再加上那些算出来高关联的节点之间的连接来加速训练过程，或者是提高准确度。

想法由来
----
这个想法的由来是 ResNet，相比 FNN，跨单层的 ResNet 网络效果更好，结构如下图所示。

![](img/0.jpg)

再到后来出现的 DenseNet 网络比 ResNet 更上一层，有着跨越多层的连接以及更加密集的连接。

![](img/1.jpg)

到这里，很明显的能看出，相比传统的层与层之间依次连接的全连接网络，不同层之间的跨层连接会更加有效。然而，像 DenseNet 那种如此密集的跨层连接感觉太浪费了，能否有一种方法让我们只连接极少数的跨层连接来达到相同的效果。为了找出那些跨层连接，这里使用 DCA 来计算不同层节点之间的关联（本人表示很牵强），选择关联强的节点进行连接。

网络实现
----
在网上去搜有没有已有的或者是类似的，但是没有找到（可能是我不会找）。Tensorflow 应该能实现这样的网络，但是我只会使用已有的模块。心想这么麻烦还不如自己动手写个简单的来的省事，因此这里再回顾下神经网络的训练过程和计算方法。

神经网络的训练过程主要是使用梯度下降（Gradient Descent）方法，经过 Loss 函数计算误差通过链式法则（Chain Rule）反向传递（Backpropagation）梯度来更新权重。

![](img/2.png)

现在我们来单看一个节点，如下图所示，计算权重 $w$ 的梯度公示为 $\frac{\partial C}{\partial w} = \frac{\partial z}{\partial w} \frac{\partial C}{\partial z}$，其中第一部分 $\frac{\partial z}{\partial w} = x$ 对应的是前向传播（Forward pass）过程，需要算出每个节点的值 $x$。$\frac{\partial C}{\partial z} = \sigma^{'}(z)\left[ w_3\frac{\partial C}{\partial z^{'}} + w_4\frac{\partial C}{\partial z^{''}} \right]$ 对应的是反向传播（Backward pass）过程，依次从后向前计算 $\frac{\partial C}{\partial z}$。

![](img/3.png)

![](img/4.png)

![](img/5.png)

总结一下上面的步骤

**定义:** 设该网络有 $n$ 层，$A_k$ 为第 $k$ 层的节点值，则有
$$ A_k = \sigma(Z_k) = \sigma(W_kA_{k-1}), \qquad A_0 = X = \left[ \begin{array}{c} x_1 \\ x_2 \\ ... \\ x_a \end{array} \right], \qquad A_n = Y = \left[ \begin{array}{c} y_1 \\ y_2 \\ ... \\ y_b \end{array} \right]  $$
其中 $\sigma$ 为激活函数，$W_k$ 为第 $k-1$ 层到第 $k$ 层的权重矩阵。这里不考虑 bias。Loss 函数为 $L(w) = f(Y, \hat{Y})$，目标是 $w^* = \arg\min\limits_{w} L(w)$，通过梯度下降更新权重 $w^t \leftarrow w^{t-1} - \eta\nabla L(w)$。

首先计算前向传播

$$
\begin{align}
Z_k &= W_k A_{k-1} \\
A_k &= \sigma(Z_k),\qquad k=1 \rightarrow n
\end{align}
$$

然后计算反向传播

$$
\begin{align}
\frac{\partial L}{\partial Z_n} &= \frac{\partial Y}{\partial Z_n}\frac{\partial L}{\partial Y} = \sigma^{'}(Z_n)\frac{\partial L}{\partial A_n} \\
\frac{\partial L}{\partial W_n} &= A_{n-1} \sigma^{'}(Z_n) \frac{\partial L}{\partial A_n}, \qquad k = n \\
\\
\frac{\partial L}{\partial Z_k} &= \frac{\partial A_k}{\partial Z_k}\frac{\partial L}{\partial A_k} \\
&= \sigma^{'}(Z_k) W_{k+1}^\mathrm{T} \frac{\partial L}{\partial Z_{k+1}} \\
\frac{\partial L}{\partial W_k} &= \frac{\partial Z_k}{\partial W_k} \frac{\partial L}{\partial Z_k} \\
&= A_{k-1}^\mathrm{T}\frac{\partial L}{\partial Z_k}, \qquad k = n-1 \rightarrow 1
\end{align}
$$

最后更新权重

$$
W^t_k = W^{t-1}_k - \eta\frac{\partial L}{\partial W_k}, \qquad k = 1\rightarrow n
$$

如果要加上任意的跨层连接，只需要修改前向传播里的

$$
Z_k = W_k A_{k-1} + \sum^{k-2}_{i=0} \sigma(i,k)A_i
$$

其中 $\sigma(i, k)$ 表示为第 i 层与第 k 层节点间是否连接的矩阵，同理修改反向传播里的

$$
\frac{\partial L}{\partial A_k} = W_{k+1}^\mathrm{T} \frac{\partial L}{\partial Z_{k+1}} + \sum^n_{i=k+2}\sigma(k,i)\frac{\partial L}{\partial Z_i}
$$

## 跨层连接网络实现

```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from skip_layer_fnn_numpy import *
```

## 效果验证

导入 MNIST 测试数据集

```python
train_data, test_data = tf.keras.datasets.mnist.load_data()
train_x, train_y = train_data
test_x, test_y = test_data

train_x_25 = []
train_y_25 = []
for i in range(train_x.shape[0]):
    if train_y[i] in (2, 5):
        train_x_25.append(train_x[i])
        train_y_25.append(train_y[i])

test_x_25 = []
test_y_25 = []
for i in range(test_x.shape[0]):
    if test_y[i] in (2, 5):
        test_x_25.append(test_x[i])
        test_y_25.append(test_y[i])


train_x_25 = (np.array(train_x_25) / 255).reshape(-1, 28*28).T
train_y_25 = np.array(np.array(train_y_25) == 2, dtype=int).reshape(-1, 1).T

test_x_25 = (np.array(test_x_25) / 255).reshape(-1, 28*28).T
test_y_25 = np.array(np.array(test_y_25) == 2, dtype=int).reshape(-1, 1).T

print("Train data: %d - label 2: %.1f%%, 5: %.1f%%" % (train_y_25.shape[1],
                                                 np.sum(train_y_25 == 1) / train_y_25.shape[1] * 100,
                                                 np.sum(train_y_25 == 0) / train_y_25.shape[1] * 100))
print("Test data: %d - label 2: %.1f%%, 5: %.1f%%" % (test_y_25.shape[1],
                                                 np.sum(test_y_25 == 1) / test_y_25.shape[1] * 100,
                                                 np.sum(test_y_25 == 0) / test_y_25.shape[1] * 100))
```

设置网络架构和参数

```python
nn_architecture = [
    {"input_dim": 784, "output_dim": 400, "activation": "sigmoid", "bias": False},
    {"input_dim": 400, "output_dim": 200, "activation": "sigmoid", "bias": False},
    {"input_dim": 200, "output_dim": 100, "activation": "sigmoid", "bias": False},
    {"input_dim": 100, "output_dim": 1, "activation": "sigmoid", "bias": False}
]

learning_rate = 0.1
```

## 训练网络 - 不使用跨层连接

```python
max_train_sample = 11000
max_train_epoch = 500
train_results = []

for sample in range(100, max_train_sample+1, 100):
    params_values = init_layers(nn_architecture)
    for epoch in range(5, max_train_epoch+1, 5):
        print("[*] sample: %d, epoch: %d" % (sample, epoch))
        epoch_n = 5
        params_values, cost_history, accuracy_history = train(train_x_25[:, :sample], train_y_25[:, :sample], nn_architecture, epoch_n, learning_rate, params_values=params_values)
        
        # test
        Y_hat, _ = full_forward_propagation(test_x_25, params_values, nn_architecture)
        cost = loss(Y_hat, test_y_25)
        accuracy = get_accuracy_value(Y_hat, test_y_25)
        
        train_results.append([sample, epoch, accuracy, cost_history[-1], accuracy_history[-1]])
```

训练时间较长，这里可以直接导入已经算好的结果

## 加入随机跨层连接

现在来看看随机添加跨层连接的效果如何

```python
def random_connects(nn_architecture, top_n):
    layer_node_num = [nn_architecture[0]['input_dim']]
    for layer in nn_architecture:
        layer_node_num.append(layer_node_num[-1] + layer['output_dim'])
   
    skip_layer_connects = {}
    n = 0
    while n < top_n:
        a = np.random.randint(layer_node_num[-1])
        b = np.random.randint(layer_node_num[-1])
        
        for i in range(len(layer_node_num)):
            if a < layer_node_num[i]:
                L1 = i
                break
        
        for i in range(len(layer_node_num)):
            if b < layer_node_num[i]:
                L2 = i
                break
        
        if L2 - L1 >= 2:
            N1 = a - (layer_node_num[L1-1] if L1 > 0 else 0)
            N2 = b - (layer_node_num[L2-1] if L2 > 0 else 0)
            if skip_layer_connects.get((L1, L2), []):
                skip_layer_connects[(L1, L2)].append((N1, N2))
            else:
                skip_layer_connects[(L1, L2)] = [(N1, N2)]
            n += 1
    
    return skip_layer_connects
```

```python
random_skip_train_results = []
pre_train_epoch = 5

for sample in range(100, max_train_sample+1, 100):
    params_values, cost_history, accuracy_history = train(train_x_25[:, :sample], train_y_25[:, :sample], nn_architecture, pre_train_epoch, learning_rate)
    
    top_n = 100
    skip_layer = random_connects(nn_architecture, top_n)
    
    for epoch in range(pre_train_epoch+5, max_train_epoch+1, 5):
        print("[*] sample: %d, epoch: %d" % (sample, epoch))
        epoch_n = 5
        params_values, cost_history, accuracy_history = train(train_x_25[:, :sample], train_y_25[:, :sample], nn_architecture, epoch_n, learning_rate, skip_layer=skip_layer, params_values=params_values)
        
        # test
        Y_hat, _ = full_forward_propagation(test_x_25, params_values, nn_architecture, skip_layer=skip_layer)
        cost = loss(Y_hat, test_y_25)
        accuracy = get_accuracy_value(Y_hat, test_y_25)
        
        skip_train_results.append([sample, epoch, accuracy, cost_history[-1], accuracy_history[-1]])
```

## 总结

最后来个总结，加入跨层连接可以加速训练过程，而且相比至于随机加入的跨层连接则没有任何效果，但是用 DCA 算节点之间的关联所花的时间太长，已经足够训练好整个模型了，而且这还是一个这么小的网络，如果有更多层或者是更多的节点，可能都算不下去，所以感觉有点得不偿失。
